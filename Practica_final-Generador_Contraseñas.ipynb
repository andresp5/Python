```python
import random
import string
from cryptography.fernet import Fernet
key = Fernet.generate_key()
fernet = Fernet(key)

aplicaciones = ["Tinder", "Instagram", "Twitter", "Tiktok"]
passwords = ["Hola", "Andres","20051206Correo!", "qwerasdf1234"]

#Funcion que define el menú
def menu ():
    accion_deseada = input("Elija una opción: \n"
                            "1- Agregar contraseña \n"
                            "2- Buscar una aplicación \n"
                            "3- Deshacer una contraseña \n"
                            "4- Modificar una contraseña \n"
                            "5- Salir del programa \n"
                            "Elección:  ")
    return accion_deseada

#Función que define la contraseña aleatoria
def agregar_password_aleatoria():
    letras = string.ascii_letters
    caracteres_especiales = string.punctuation
    alfabeto = letras + caracteres_especiales
    new_password = ''.join([random.choice(alfabeto) for i in range(20)])
    return new_password

#Función para encriptar la contraseña
def encriptar_password(password):

    mensaje_enc = fernet.encrypt(password.encode())

    return mensaje_enc

#Función para desencriptar la contraseña
def desencriptar_password(mensaje_enc):

    password = fernet.decrypt(mensaje_enc.decode())

    return password

#Función para agregar contraseña
def agregar_password ():
    aplicacion_usuario = input("Dime una aplicación: ")
    if aplicacion_usuario not in aplicaciones:
        aplicaciones.append(aplicacion_usuario)#introduce aplicaciones a la lista aplicaciones
        while True:
            password_usuario = input("Dime una contraseña o dime 'YES' si la quieres aleatoria : ")
            if password_usuario not in passwords and password_usuario.upper() != "YES": #introduce la contraseña en la lista de contraseñas
                encriptacion = input("Desea que sea encriptada?Y/N: ").upper() #encriptación de la contraseña
                if encriptacion == "Y":
                    passwords.append(encriptar_password(password_usuario))
                    return
                else:
                    passwords.append(password_usuario)
                    return
            elif password_usuario.upper() == "YES": #Introducción de la contraseña aleatoria en la lista de contraseñas
                encriptacion = input("Desea que sea encriptada?Y/N: ").upper() #encriptación de la contraseña
                if encriptacion == "Y":
                    passwords.append(encriptar_password(agregar_password_aleatoria()))
                    return
                else:
                    passwords.append(agregar_password_aleatoria())
                    return
            else:
                print("La contraseña ya existe")
    else:
        print("La aplicación ya está creada")
        return


#Función para buscar una aplicación
def buscar_una_aplicacion ():

    aplicacion_usuario = input("Dime una aplicación: ")

    for n in range(len(aplicaciones)):
        if aplicaciones[n] == aplicacion_usuario:
            if len(passwords[n]) > 25: #Cuenta el largo para determinar si esta encriptada
              print(f"La contraseña de tu aplicación es '{desencriptar_password(passwords[n])}'")
              return
            else:
              print(f"La contraseña de tu aplicación es '{passwords[n]}'")
              return
    else:
        print("La contraseña de esa aplicación no está creada")

#Función para deshacer la ultima aplicación y contraseña
def deshacer_la_ultima():

    if len(aplicaciones) > 4:
      aplicaciones.pop()
      passwords.pop()
      print("Ultima aplicación y contraseña eliminadas")
      return
    else:
      print("No se pueden eliminar las contraseñas existentes")
      return

#Función para modificar la contraseña
def modificar_password():

    aplicacion_usuario = input("Dime la aplicación de la que quieres modificar la contraseña: ")

    for n in range(len(aplicaciones)):
        if aplicaciones[n] == aplicacion_usuario:
            passwords.remove(passwords[n])
            while True:
                password_usuario = input("Dime una contraseña o dime 'YES' si la quieres aleatoria : ")
                if password_usuario not in passwords and password_usuario.upper() != "YES":
                  encriptacion = input("Desea que sea encriptada?Y/N: ").upper() #encriptación de la contraseña
                  if encriptacion == "Y":
                    passwords.insert(n, encriptar_password(password_usuario))
                    return
                  else:
                    passwords.insert(n, password_usuario)
                    return
                elif password_usuario.upper() == "YES": #Introducción de la contraseña aleatoria en la lista de contraseñas
                    encriptacion = input("Desea que sea encriptada?Y/N: ").upper() #encriptación de la contraseña
                    if encriptacion == "Y":
                      passwords.insert(n, encriptar_password(agregar_password_aleatoria()))
                      return
                    else:
                      passwords.insert(n, agregar_password_aleatoria())
                      return
                else:
                    print("La contraseña ya existe")

    else:
        print("La contraseña de esa aplicación no está creada")
        return


while True:
    print("\n")
    opcion_deseada = menu()
    if opcion_deseada == "1":
        agregar_password()
    elif opcion_deseada == "2":
        buscar_una_aplicacion()
    elif opcion_deseada == "3":
        deshacer_la_ultima()
    elif opcion_deseada == "4":
        modificar_password()
    elif opcion_deseada == "5":
        salida = input("Desea salir del programa? Y/N").upper()
        if salida == "Y":
            break
    else:
        print("Esa opción no es válida")
```
