```python
from random import choice

palabras = ["diccionario", "avioneta","circuito", "altavoces", "concierto", "vagabundo", "jabalina", "alzamiento"]
intentos = 6
aciertos = 0
letras_correctas = []
letras_incorrectas = []
juego_terminado = False

def elegir_palabra(lista_palabras):
    palabra_elegida = choice(lista_palabras)
    letras_unicas = len(set(palabra_elegida))

    return palabra_elegida, letras_unicas

""" La función set se utiliza para contar las letras repetidas que hay en palabra
Como la función set elimina los duplicados, utilizamos esa función y despues le añadimos
la función len para que sume los elementos eliminados"""

def pedir_letra():
    letra_elegida = ""
    es_valida = False
    abecedario = "abcdefghijklmnñopqrstvwxyz"

    while not es_valida:
        letra_elegida = input("Elige una letra: ").lower()

        if letra_elegida in abecedario and len(letra_elegida) == 1:
            es_valida = True
        else:
            print("No has elegido la letra correcta")

    return letra_elegida
""" El loop esta diciendo que mientras la condición de False no cambie en la variable,
seguirá pidiendo la letra. Para que cambie la condición le ponemos un condicional 
que diga si la letra elegida no esta en el abecedario y es igual que 1 puedes considerar
la variable como True. Si no pasa eso, imprime que no lo has hecho bien"""

def mostrar_nuevo_tablero(palabra_elegida):

    lista_oculta = []

    for l in palabra_elegida:
        if l in letras_correctas:
            lista_oculta.append(l)
        else:
            lista_oculta.append("-")

    print(" ".join(lista_oculta))

"""Para cada letra de la palabra elegida, si la letra esta en letras correctas, 
añade a la lista oculta esa letra. Si no aparece añade un guion
Se crea un loop porque necesitamos que para cada introducción de una letra nueva
nos muestre el estado de la palabra"""

def chequear_letra(letra_elegida, palabra_oculta, vidas, coincidencias):

    fin = False

    if letra_elegida in palabra_oculta and letra_elegida not in letras_correctas:
        letras_correctas.append(letra_elegida)
        coincidencias += 1
    elif letra_elegida in palabra_oculta and letra_elegida in letras_correctas:
        print("Ya has acertado esa letra, prueba con otra distinta")
    else:
        letras_incorrectas.append(letra_elegida)
        vidas -= 1
    if vidas == 0:
        fin = perder()
    elif coincidencias == letras_unicas:
        fin = ganar(palabra_oculta)

    return vidas, fin, coincidencias

"""En esta función esta codicionando que si la letra elegida esta dentro de la palabra
oculta que lo sume a la variable de letras correctas y a la de coincidencias, en caso contrario
a la de incorrectas y que le reste una vida
Despues esta condicionando que si no le quedan vidas ejecute la funcion perder o 
si las coincidencias son iguales al numero de letras que ejecute la funcion ganar"""

def perder():

    print("Te has quedado sin vidas")
    print("La palabra correcta era " + palabra)

    return True

def ganar(palabra_descubierta):

    mostrar_nuevo_tablero(palabra_descubierta)
    print("Felicidades,has encontrado la palabra")

    return True

palabra, letras_unicas = elegir_palabra(palabras)

while not juego_terminado:
    print("\n" + "*" * 20 + "\n")
    mostrar_nuevo_tablero(palabra)
    print("\n")
    print("Letras incorrectas: " + "-".join(letras_incorrectas))
    print(f"Vidas: {intentos}")
    print("\n" + "*" * 20 + "\n")

    letra = pedir_letra()

    intentos, terminado, aciertos = chequear_letra(letra, palabra, intentos, aciertos)

    juego_terminado = terminado
```
